```python
import pandas as pd
import numpy as np 
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('monster_com.csv')
```

```python
df.head()
```

```python
def inspect_dataframe(input_df):
    
    # Basic Info
    print("Basic Info:")
    print(input_df.info())
    print("-" * 50)
    
    # Missing Values
    print("Missing Values Count:")
    print(input_df.isnull().sum())
    print("-" * 50)
   
    # Missing Values Percentage
    print("\nMissing Values Percentage:")
    print((input_df.isnull().sum() / len(input_df)) * 100)
    print("-" * 50)
    
    # Summary Statistics
    print("Summary Statistics:")
    print(input_df.describe())
    print("-" * 50)
    
    #Duplicates
    print("Duplicates present:")
    print(input_df.duplicated().sum())
    
    return "Inspection Done"
    
    


inspect_dataframe(df)
```

```python
columns_to_drop = ['date_added', 'salary', 'organization', 'country_code','country','has_expired','page_url','uniq_id','job_board']
cleaned_df = df.drop(columns=columns_to_drop)
```

```python
cleaned_df.head()
```

```python
inspect_dataframe(cleaned_df)
```

```python
cleaned_df.dropna(subset=['sector'], inplace=True)
cleaned_df.dropna(subset=['job_type'], inplace=True)
```

```python
cleaned_df.head()
```

```python
inspect_dataframe(cleaned_df)
```

```python
cleaned_df.to_csv('cleaned_mons')
```
